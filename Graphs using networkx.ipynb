# GRAPHS

```python
import networkx as nx
```

### BARBELL GRAPH

```python
G=nx.barbell_graph(4,2)
nx.draw(G)
```

```python
G=nx.barbell_graph(5,3)
nx.draw(G,node_color="lightcoral")
```

### COMPLETE GRAPHS

```python
G=nx.complete_graph(4)
nx.draw(G,node_color="red")
```

```python
G=nx.complete_graph(5)
nx.draw(G,node_color="green")
```

### CYCLE GRAPHS

```python
G=nx.cycle_graph(6)
nx.draw(G,node_color="yellow")
```

```python
G=nx.cycle_graph(8)
nx.draw(G,node_color="purple")
```

### LADDER GRAPH

```python
G=nx.ladder_graph(5)
nx.draw(G,node_color="pink")
```

```python
G=nx.ladder_graph(7)
nx.draw(G,node_color="black")
```
